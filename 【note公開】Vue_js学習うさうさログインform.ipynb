````python
from IPython.core.display import display, HTML  # HTMLをColab上で表示するためのimport

display(HTML('''
<!DOCTYPE html>
<html lang="ja">
<head>
  <meta charset="UTF-8">
  <title>うさうさパン屋さん ログインフォーム</title>
  <script src="https://cdn.jsdelivr.net/npm/vue@2"></script>
  <style>
    body { font-family: sans-serif; margin: 20px; }
    .error { color: red; }
    .success { color: green; }
    input { margin: 5px 0; padding: 5px; width: 300px; }
    button { padding: 6px 12px; margin-top: 10px; }
    .welcome-box { border: 1px solid #ccc; padding: 15px; border-radius: 8px; background: #f8f8f8; }
  </style>
</head>
<body>

<div id="app">
  <h2 v-if="!isLoggedIn">うさうさパン屋さん ログインフォーム</h2>

  <!-- ログイン画面 -->
  <div v-if="!isLoggedIn">
    <label>メールアドレス:</label><br>
    <input type="text" v-model="email" placeholder="メールを入力してください"><br>

    <label>パスワード:</label><br>
    <input type="password" v-model="password" placeholder="パスワードを入力してください"><br>

    <button @click="login">ログイン</button>

    <div v-if="errorMessages.length" class="error">
      <p><strong>入力エラー:</strong></p>
      <ul>
        <li v-for="(err, index) in errorMessages" :key="index">{{ err }}</li>
      </ul>
    </div>

    <div v-if="successMessage" class="success">
      {{ successMessage }}
    </div>
  </div>

  <!-- ログイン成功後の画面 -->
  <div v-if="isLoggedIn" class="welcome-box">
    <h3>ログイン成功！ようこそ、{{ email }} さん🐰</h3>
    <p>今日もたくさんパンを売りましょう！🥐🥖🍞</p>
    <button @click="logout">ログアウト</button>
  </div>
</div>

<script>
new Vue({
  el: '#app',
  data: {
    email: '',
    password: '',
    errorMessages: [],
    successMessage: '',
    isLoggedIn: false  // ログイン状態を管理
  },
  methods: {
    login() {
      this.errorMessages = [];
      this.successMessage = '';

      // バリデーション
      if (!this.email) {
        this.errorMessages.push("メールアドレスを入力してください。");
      } else if (!this.validEmail(this.email)) {
        this.errorMessages.push("正しいメールアドレス形式で入力してください。");
      }

      if (!this.password) {
        this.errorMessages.push("パスワードを入力してください。");
      } else if (this.password.length < 6) {
        this.errorMessages.push("パスワードは6文字以上必要です。");
      }

      // 正しい認証情報（仮のアカウント）
      const validEmail = "usa@bakery.jp";
      const validPassword = "usapan123";

      if (this.errorMessages.length === 0) {
        if (this.email === validEmail && this.password === validPassword) {
          this.isLoggedIn = true;
          this.successMessage = "";
        } else {
          this.errorMessages.push("メールアドレスまたはパスワードが間違っています。");
        }
      }
    },
    logout() {
      this.isLoggedIn = false;
      this.email = '';
      this.password = '';
      this.successMessage = '';
    },
    validEmail(email) {
      const re = /^[^\\s@]+@[^\\s@]+\\.[^\\s@]+$/;
      return re.test(email);
    }
  }
});
</script>

</body>
</html>
'''))
````

## 🧭 コードの全体の流れ

---

### 🔹1. Google Colab 上で Vue.js を表示する準備
```python
from IPython.core.display import display, HTML
```
→ `IPython.display` を使って、HTMLをGoogle Colabのセル内に表示できるようにします。

---

### 🔹2. Vue.js + HTML + CSS を `display(HTML(...))` で表示

- `<!DOCTYPE html>` から始まる HTML を Colab に埋め込みます。
- 外部から Vue.js (CDN) を読み込むため `<script src="...">` を使います。
- ログインフォームのデザインを簡単なCSSで装飾。

---

### 🔹3. フォームの見た目と条件分岐
```html
<div v-if="!isLoggedIn">...</div>
```
→ Vue の `v-if` を使って、ログイン前のフォームとログイン後の画面を切り替えます。

---

### 🔹4. 入力データを Vue 側の変数にバインド
```html
<input type="text" v-model="email">
<input type="password" v-model="password">
```
→ 入力された「メールアドレス」と「パスワード」は Vue の `data` 内の変数にリアルタイムで格納されます。

---

### 🔹5. ログインボタンが押されたら `login()` 関数が実行
```html
<button @click="login">ログイン</button>
```
→ `@click` はイベントリスナーで、クリックされたときに `methods.login()` を実行。

---

### 🔹6. `login()` 関数の中身（Vueの methods 内）

1. エラーメッセージを初期化
2. 入力バリデーション（空欄チェック、メール形式チェック、パスワード長さ）
3. 正しいアカウント（今回は仮に `"usa@bakery.jp"` / `"usapan123"`）と一致しているかを判定
4. 一致していれば `isLoggedIn = true` にする（ログイン成功）
5. 一致していなければエラーメッセージを表示

---

### 🔹7. ログイン成功後の画面
```html
<div v-if="isLoggedIn" class="welcome-box">...</div>
```
→ `isLoggedIn` が `true` になったら、ログイン成功メッセージと「ログアウト」ボタンが表示されます。

---

### 🔹8. ログアウトボタンで状態リセット
```html
<button @click="logout">ログアウト</button>
```
→ 変数を初期化してログアウト状態に戻します。
